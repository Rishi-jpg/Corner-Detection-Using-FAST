# Importing library

```python
# organizing imports 
import cv2 
from google.colab.patches import cv2_imshow
import numpy as np 
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

# Showing Original Image

```python
# path to input image specified and image is loaded with imread command  
input_path = r'input/clouds.jpeg'
image = cv2.imread(input_path, cv2.IMREAD_COLOR) 
cv2_imshow(image)
```

# Implementing FAST Algorithm with OpenCV

```python
# Initialize the FAST object with default values
fast = cv2.FastFeatureDetector_create()

# find and draw the keypoints
image_detect = fast.detect(image,None)
# output_image = image.copy()
img_1 = cv2.drawKeypoints(image, image_detect, None, color=(255,0,0))


# Print all default params
print( "Threshold: {}".format(fast.getThreshold()) )
print( "nonmaxSuppression:{}".format(fast.getNonmaxSuppression()) )
print( "neighborhood: {}".format(fast.getType()) )
print( "Total Keypoints with nonmaxSuppression: {}".format(len(image_detect)) )

cv2.imwrite('output/with_nonmaxSuppression.png',img_1)

# Disable nonmaxSuppression 
fast.setNonmaxSuppression(0)

# find and draw the keypoints
kp = fast.detect(image,None)
print( "Total Keypoints without nonmaxSuppression: {}".format(len(kp)) )
img_2 = cv2.drawKeypoints(image, kp, None, color=(255,0,0))
cv2.imwrite('output/without_nonmaxSuppression.png',img_2)
```

# See the results

```python
# img_1 shows FAST with nonmaxSuppression
cv2_imshow(img_1)

# img_2 shows FAST without nonmaxSuppression
cv2_imshow(img_2)
```
